```python
# IMPORT MODULE 
import numpy as np
import matplotlib.pyplot as plt

# Contoh fungsi & derivatif 
f = lambda x: x**3 - 4*x - 9  #  x^3 - 4x - 9
df = lambda x: 3*x**2 - 4     # Derivatif

a, b = 2, 3  # Interval 
x0, x1 = 2, 3  # Initial guess
```

### **BISECTION METHOD**


Metode bisection digunakan untuk menemukan akar suatu fungsi dengan cara membagi interval menjadi dua bagian.
Ini sangat berguna ketika solusi eksak tidak dapat ditemukan secara aljabar, terutama untuk fungsi kompleks atau tak terdiferensial.
Metode ini sederhana, stabil, dan selalu konvergen jika ada perubahan tanda dalam interval yang diberikan.


```python
def bisection_method(f, a, b, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        c = (a + b) / 2
        print(f"Iterasi {i+1}: c = {c}")
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c


root_bisection = bisection_method(f, a, b)
print("Akar (Bisection):", root_bisection)

# Plot Bisection Method
x_vals = np.linspace(1, 4, 400)
y_vals = f(x_vals)
plt.figure()
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 4x - 9')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter([root_bisection], [0], color='r', label='Bisection Root')
plt.legend()
plt.title("Bisection Method")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### **Metode Newton-Raphson**

Metode Newton-Raphson menggunakan turunan pertama dari fungsi untuk mempercepat konvergensi menuju akar.

```python
def newton_raphson_method(f, df, x0, tol=1e-6, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_new = x - f(x) / df(x)
        print(f"Iteration {i+1}: x = {x}, x_new = {x_new}")
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

# Newton-Raphson Method Results
root_newton = newton_raphson_method(f, df, x0)
print("Akar (Newton-Raphson):", root_newton)

# Plot Newton-Raphson Method
plt.figure()
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 4x - 9')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter([root_newton], [0], color='g', label='Newton Root')
plt.legend()
plt.title("Newton-Raphson Method")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### **Metode Secant**
Metode Secant adalah salah satu metode numerik untuk mencari akar dari suatu fungsi non-linear. Metode ini merupakan pengembangan dari Metode Newton-Raphson, tetapi tidak memerlukan perhitungan turunan, sehingga lebih efisien dalam beberapa kasus.

Langkah-langkah Metode Secant : 
- Pilih dua tebakan awal x0 dan x1.
- Hitung x2â€‹ menggunakan rumus secant.
- Evaluasi f(x2). Jika cukup dekat dengan nol (sesuai toleransi kesalahan), berhenti.
- Jika belum memenuhi toleransi, perbarui x1 = x2 dan x0 = x1, lalu ulangi proses hingga konvergen.
 

```python
def secant_method(f, x0, x1, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        x_new = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        print(f"Iterasi {i+1}: x_new = {x_new}")
        if abs(x_new - x1) < tol:
            return x_new
        x0, x1 = x1, x_new
    return x1

# Secant Method Results
root_secant = secant_method(f, x0, x1)
print("Akar (Secant):", root_secant)

# Plot Secant Method
plt.figure()
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 4x - 9')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter([root_secant], [0], color='b', label='Secant Root')
plt.legend()
plt.title("Secant Method")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### **Metode regula falsi**
Metode regula falsi adalah variasi dari metode bagi dua, tetapi menggunakan pendekatan garis lurus untuk memperkirakan akar.

```python
def regula_falsi_method(f, a, b, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        c = b - f(b) * (b - a) / (f(b) - f(a))
        print("Iterasi", i+1, ":", c)
        if abs(f(c)) < tol:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return c

# Regula Falsi Method Results
root_regula = regula_falsi_method(f, a, b)
print("Akar (Regula Falsi):", root_regula)

# Plot Regula Falsi Method
plt.figure()
plt.plot(x_vals, y_vals, label='f(x) = x^3 - 4x - 9')
plt.axhline(0, color='black', linewidth=0.5)
plt.scatter([root_regula], [0], color='y', label='Regula Falsi Root')
plt.legend()
plt.title("Regula Falsi Method")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.show()
```

### **Kesimpulan**  

Dari hasil implementasi dan visualisasi empat metode pencarian akar pada persamaan non-linear, diperoleh beberapa kesimpulan sebagai berikut:  

1. **Metode Bisection**  
   - Konvergen secara pasti tetapi relatif lambat.  
   - Selalu mempersempit interval yang mengandung akar.  
   - Cocok untuk kasus di mana eksistensi akar dalam suatu interval sudah diketahui.  

2. **Metode Newton-Raphson**  
   - Konvergen lebih cepat dibandingkan bisection, tetapi bergantung pada tebakan awal yang baik.  
   - Bisa gagal atau mengalami divergensi jika turunan pertama mendekati nol.  
   - Lebih efisien jika fungsi dan turunannya dapat dihitung dengan mudah.  

3. **Metode Secant**  
   - Mirip dengan Newton-Raphson tetapi tidak memerlukan turunan eksplisit.  
   - Bisa mengalami ketidakstabilan atau konvergensi lambat dalam beberapa kasus.  
   - Cocok untuk situasi di mana turunan sulit dihitung.  

4. **Metode Regula Falsi (False Position)**  
   - Kombinasi antara bisection dan secant yang tetap menjaga bracket akar.  
   - Biasanya lebih cepat dari bisection tetapi masih bisa lebih lambat dibandingkan Newton-Raphson.  
   - Bisa mengalami stagnasi jika salah satu ujung interval tetap tidak berubah.  

Secara umum, pemilihan metode tergantung pada sifat fungsi dan kondisi awal yang diberikan. Newton-Raphson dan Secant cenderung lebih cepat, tetapi Bisection dan Regula Falsi lebih aman dalam memastikan konvergensi.
