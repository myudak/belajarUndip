# GYAT KELOMPOK

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. Galat Absolut dan Relatif

```python
true_value = 100
measured_value = 95  # Misalnya terjadi kesalahan pengukuran

absolute_error = abs(true_value - measured_value)
relative_error = (absolute_error / true_value) * 100

print("Contoh: Galat Absolut dan Relatif")
print("Nilai Sebenarnya:", true_value)
print("Nilai Ukuran:", measured_value)
print("Galat Absolut:", absolute_error)
print(f"Galat Relatif: {relative_error:.2f}%\n")
```

# 2. Galat Sistematis (Kesalahan Konsisten)

```python
true_values = np.linspace(50, 150, 10)
systematic_bias = 5  # Setiap pengukuran mengalami bias +5
measured_values_systematic = true_values + systematic_bias

plt.figure(figsize=(6, 4))
plt.plot(true_values, measured_values_systematic, 'o-', label='Nilai Ukuran (Bias +5)')
plt.plot(true_values, true_values, 'r--', label='Nilai Sebenarnya')
plt.xlabel('Nilai Sebenarnya')
plt.ylabel('Nilai Ukuran')
plt.title('Galat Sistematis')
plt.legend()
plt.grid(True)
plt.show()

print("Contoh: Galat Sistematis")
print("Nilai Sebenarnya:", true_values)
print("Nilai Ukuran (Bias +5):", measured_values_systematic)
```

# 3. Galat Acak (Kesalahan Tidak Konsisten)

```python
np.random.seed(42)
random_noise = np.random.normal(0, 3, size=true_values.shape)
measured_values_random = true_values + random_noise

plt.figure(figsize=(6, 4))
plt.plot(true_values, measured_values_random, 'o', label='Nilai Ukuran (Galat Acak)')
plt.plot(true_values, true_values, 'r--', label='Nilai Sebenarnya')
plt.xlabel('Nilai Sebenarnya')
plt.ylabel('Nilai Ukuran')
plt.title('Galat Acak')
plt.legend()
plt.grid(True)
plt.show()

print("Contoh: Galat Acak")
print("Nilai Sebenarnya:", true_values)
print("Nilai Ukuran dengan Galat Acak:", measured_values_random)
```

# 4. Galat Lingkungan (Misalnya Pengaruh Suhu)

```python
ambient_temperature = np.random.uniform(-2, 2, size=true_values.shape)  # Variasi suhu yang mempengaruhi hasil
measured_values_environmental = true_values + ambient_temperature

plt.figure(figsize=(6, 4))
plt.plot(true_values, measured_values_environmental, 'o', label='Nilai Ukuran (Galat Lingkungan)')
plt.plot(true_values, true_values, 'r--', label='Nilai Sebenarnya')
plt.xlabel('Nilai Sebenarnya')
plt.ylabel('Nilai Ukuran')
plt.title('Galat Lingkungan')
plt.legend()
plt.grid(True)
plt.show()

print("Contoh: Galat Lingkungan")
print("Nilai Sebenarnya:", true_values)
print("Nilai Ukuran dengan Pengaruh Lingkungan:", measured_values_environmental)
```
